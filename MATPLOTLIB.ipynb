```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic')
```

```python
%matplotlib inline
```

```python
x = np.linspace(1,10,100)
```

```python
fig = plt.figure()

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
fig.savefig('my_figure.png')
```
